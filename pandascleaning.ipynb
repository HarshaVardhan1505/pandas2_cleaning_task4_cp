## **Missing value treatments**

### For some numerical columns which have outliers ,we are filling missing values with '**Median** or **KNN**'

```python
#flling missing values with medain
for i in ['BMI','Polio','Income Composition Of Resources']:
    df[i].fillna(df['BMI'].median(),inplace=True)
```

```python
#filling missing values with KNN
from sklearn.impute import KNNImputer
impute=KNNImputer()
```

```python
for i in df.select_dtypes(include='number').columns:
    df[i]=impute.fit_transform(df[[i]])
```

# Treating outliers

```python
#creating a copy of data
df1=df.copy(deep=True)
```

```python
#creating a function for finding quartiles
def whisker(col):
    q1,q3=np.percentile(col,[25,75])
    iqr=q3-q1
    lw=q1-1.5*iqr
    uw=q3+1.5*iqr
    return lw,uw
```

```python
for i in ['GDP', 'Population', 'Thinness  1-19 Years',
       'Thinness 5-9 Years', 'Income Composition Of Resources']:
    print(whisker(df1[i]))
```

```python
for i in ['GDP', 'Population', 'Thinness  1-19 Years',
       'Thinness 5-9 Years', 'Income Composition Of Resources']:
    lw,uw=whisker(df1[i])
    df1[i]=np.where(df1[i]<lw,lw,df1[i])
    df1[i]=np.where(df1[i]>uw,uw,df1[i])
```

```python
for i in df1[['GDP', 'Population', 'Thinness  1-19 Years',
       'Thinness 5-9 Years', 'Income Composition Of Resources']].columns: 
    sns.boxplot(data=df1,x=i)
    plt.show()
```
